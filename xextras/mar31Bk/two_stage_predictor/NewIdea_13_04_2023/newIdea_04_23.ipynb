```python
import pandas as pd

dataset = pd.read_csv('D:\\NewFinalDatasetsForNeuralNetwork\\New_Encap_25_50_75_NoCrash_01_99_crash_noncrash_combined_csv_24_03_2023__11_02.csv')
dataset
```

```python
dataset = dataset.dropna()
dataset = dataset.drop_duplicates()
dataset = dataset.drop(dataset[dataset['TTC'] < 0].index)
dataset
```

```python
import numpy as np

# Create a new column 'Scenario type' and map values to 'crash' or 'no_crash' based on their TTC range
bins = [-1, 7, np.inf]
labels = ['crash', 'no_crash']
dataset['SCENARIO TYPE'] = pd.cut(dataset['TTC'], bins=bins, labels=labels)
dataset
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select column to one-hot encode
col_to_encode = 'SCENARIO TYPE'

# Perform one-hot-encoding
one_hot_encoded = pd.get_dummies(dataset[col_to_encode], prefix=col_to_encode)

# Append one-hot encoded columns to original dataset
dataset = pd.concat([dataset, one_hot_encoded], axis=1)

# Drop original column that was one-hot encoded
dataset = dataset.drop(col_to_encode, axis=1)
dataset
```

```python
# Split dataset into features and labels
X = dataset.drop(columns=["TTC", "SCENARIO TYPE_crash", "SCENARIO TYPE_no_crash"]).values
y_regression = dataset[["TTC"]].values
y_classification = dataset[["SCENARIO TYPE_crash", "SCENARIO TYPE_no_crash"]].values
n_features = X.shape[1]
print(X)
print(n_features)
print(y_classification)
print(y_regression)
```

```python
from numpy import unique

n_class = len(unique(y_classification))
n_class
```

```python
from sklearn.model_selection import train_test_split
# Split dataset into training and testing sets
X_train, X_test, y_train_classification, y_test_classification, y_train_regression, y_test_regression  = train_test_split(X, y_classification, y_regression, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.layers import Input
from keras.layers import Dense, Normalization

class MyModel(keras.Model):
    def __init__(self, n_features):
        super(MyModel, self).__init__()
        visible = Input(shape=(n_features,))
        self.normalization = layers.Normalization(input_shape= (n_features,), axis=None)
        #self.normalization.adapt(np.array(X_train))
        self.dense1 = layers.Dense(128, activation='relu')
        self.dense2 = layers.Dense(64, activation='sigmoid')
        self.dense3 = layers.Dense(32, activation='relu')
        self.classification_output = layers.Dense(n_class, activation='softmax')
        
        self.dense4 = layers.Dense(64, activation='relu')
        #model.add(BatchNormalization())
        self.dense5 = layers.Dense(128, activation='relu')
        self.dense6 = layers.Dense(64, activation='relu')
        self.regression_output = layers.Dense(1)
        
    def call(self, inputs):
        # Normalize the inputs
        normalized_inputs = self.normalization(inputs)

        # Pass the inputs through the classification layers
        y = self.dense1(normalized_inputs)
        y = self.dense2(y)
        y = self.dense3(y)
        classification_output = self.classification_output(y)

        # Check if the classification output is 0
        is_zero = tf.equal(tf.argmax(classification_output, axis=1), 0)

         # Define the computation for the two cases
        def regression_case():
        # Pass the original inputs through the regression layers
            x = self.dense4(normalized_inputs)
            x = self.dense5(x)
            x = self.dense6(x)
            return self.regression_output(x)

        def classification_case():
            return classification_output

        # Conditionally execute the appropriate computation
        output = tf.cond(is_zero, regression_case, classification_case)
        return output
```
