```python
! pip install graphviz pydot
import tensorflow as tf
import keras
import numpy as np
from keras.utils import plot_model


# Defining first layer of the neural network, which normalizes the input data on the fly
normalizer_layer = keras.layers.Normalization(axis=-1)
normalizer_layer = keras.layers.Normalization(input_shape=[4,], axis=None)

# Adapting normalizer layer to the input train data shape
normalizer_layer.adapt(np.array(X_train))
class_model = nn_model = keras.Sequential([
      normalizer_layer,
      keras.layers.Dense(128, activation='relu', kernel_initializer='he_uniform', name = 'Hidden_Layer1'),
      keras.layers.Dense(64, activation='sigmoid',name = 'Hidden_Layer2'),
      keras.layers.Dense(32, activation='relu',name = 'Hidden_Layer3'),
      #keras.layers.Dense(32, activation='sigmoid'),
      #keras.layers.Dense(128, activation='sigmoid'),
      keras.layers.Dense(y.shape[1], activation="softmax", name = 'Output_Layer')
  ])

# Assume 'model' is a trained neural network model
plot_model(class_model, to_file='C:\\Users\\thiru\\OneDrive\\Desktop\\Aashritha project\\NeuralNetModel Structures\\classification_model_diagram.png', show_shapes=True, show_layer_names=True)
```
